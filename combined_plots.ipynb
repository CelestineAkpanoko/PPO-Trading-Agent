```python
import numpy as np
import matplotlib.pyplot as plt

def plot_combined_performance(rewards, total_losses, portfolio_returns, figure_file):
    x = np.arange(len(rewards))  # Assuming all arrays are of the same length

    plt.figure(figsize=(12, 6))

    # Normalizing data for combined plot
    normalized_rewards = rewards / np.max(np.abs(rewards))
    normalized_losses = total_losses / np.max(np.abs(total_losses))
    normalized_portfolio_returns = portfolio_returns / np.max(np.abs(portfolio_returns))

    # Plotting Rewards, Losses, and Portfolio Returns
    plt.plot(x, normalized_rewards, label='Normalized Rewards', color='blue')
    plt.plot(x, normalized_losses, label='Normalized Total Loss', color='red')
    plt.plot(x, normalized_portfolio_returns, label='Normalized Portfolio Return', color='green')

    plt.title('Combined Performance Over Episodes')
    plt.xlabel('Episode')
    plt.ylabel('Normalized Values')
    plt.legend()

    plt.tight_layout()
    plt.savefig(figure_file)
    plt.show()
```

```python
# Load the data
rewards = np.load('rewards.npy')
total_losses = np.load('total_losses.npy')
portfolio_returns = np.load('portfolio_returns.npy')
plot_combined_performance(rewards, total_losses, portfolio_returns, 'combined_performance_plot.png')
```
